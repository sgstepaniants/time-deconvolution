```python
import math
import numpy as np
import scipy
import matplotlib.pyplot as plt

from probability import *
from hilbert_transform import *
from example_measures import lmbda_1, lmbda_2, lmbda_3, lmbda_4

plt.rc('text', usetex=True)
plt.rc('font', family='serif', size=15)
import matplotlib
matplotlib.rcParams['text.latex.preamble'] = r"\usepackage{amsmath}"
```

```python
Hlmbda_1 = HilbertTransform(lmbda_1)
Hlmbda_1_analytic = lambda x: 1/(math.pi) * np.sum(lmbda_1.atom_wts / (x[:, None] - lmbda_1.atoms), axis=1)

fig, ax = plt.subplots(figsize=(5, 5))
ax.set_title("Distribution $\lambda_1$")
if lmbda_1.density is not None:
    ax.plot(lmbda_1.quad_pts, lmbda_1.density_vals, color="blue")
ax.scatter(lmbda_1.atoms, lmbda_1.atom_wts, marker='o', color="blue")
ax.vlines(lmbda_1.atoms, ymin=0, ymax=lmbda_1.atom_wts, color="blue")
ax.set_xlim(-10, 10)
plt.show()

H_comp = Hlmbda_1(lmbda_1.quad_pts)
H_analytic = Hlmbda_1_analytic(lmbda_1.quad_pts)
rel_err = np.linalg.norm(H_comp - H_analytic) / np.linalg.norm(H_analytic)

fig, ax = plt.subplots(figsize=(5, 5))
ax.set_title(f"Hilbert Transform $H[\lambda_1]$ (Rel. Err. {rel_err})")
ax.plot(lmbda_1.quad_pts, H_comp, color="blue")
ax.plot(lmbda_1.quad_pts, H_analytic, color="k", linestyle='dashed')
ax.set_xlim(-10, 10)
#ax.set_yscale("symlog")
ax.legend(["Computed", "Analytic"])
plt.show()
```

```python
Hlmbda_2 = HilbertTransform(lmbda_2)
a = 1
Hlmbda_2_analytic = lambda x: np.sign(x)/(math.pi) * (np.exp(-a*np.abs(x))*scipy.special.expi(a*np.abs(x)) - np.exp(a*np.abs(x))*scipy.special.expi(-a*np.abs(x)))

fig, ax = plt.subplots(figsize=(5, 5))
ax.set_title("Distribution $\lambda_2$")
if lmbda_2.density is not None:
    ax.plot(lmbda_2.quad_pts, lmbda_2.density_vals, color="blue")
ax.scatter(lmbda_2.atoms, lmbda_2.atom_wts, marker='o', color="blue")
ax.vlines(lmbda_2.atoms, ymin=0, ymax=lmbda_2.atom_wts, color="blue")
ax.set_xlim(-10, 10)
plt.show()

H_comp = Hlmbda_2(lmbda_2.quad_pts)
H_analytic = Hlmbda_2_analytic(lmbda_2.quad_pts)
rel_err = np.linalg.norm(H_comp - H_analytic) / np.linalg.norm(H_analytic)

fig, ax = plt.subplots(figsize=(5, 5))
ax.set_title(f"Hilbert Transform $H[\lambda_2]$ (Rel. Err. {rel_err:.2e})")
ax.plot(lmbda_2.quad_pts, H_comp, color="blue")
ax.plot(lmbda_2.quad_pts, H_analytic, color="k", linestyle='dashed')
ax.set_xlim(-10, 10)
#ax.set_yscale("symlog")
ax.legend(["Computed", "Analytic"])
plt.show()
```

```python
Hlmbda_3 = HilbertTransform(lmbda_3)
a = 1
H_parabolic = lambda x: 1/math.pi * (2*x/a - (1 - (x/a)**2)*np.log(np.abs((x-a)/(x+a))))
Hlmbda_3_analytic = lambda x: H_parabolic(x - 3) + H_parabolic(x + 3)

fig, ax = plt.subplots(figsize=(5, 5))
ax.set_title("Distribution $\lambda_3$")
if lmbda_3.density is not None:
    ax.plot(lmbda_3.quad_pts, lmbda_3.density_vals, color="blue")
ax.scatter(lmbda_3.atoms, lmbda_3.atom_wts, marker='o', color="blue")
ax.vlines(lmbda_3.atoms, ymin=0, ymax=lmbda_3.atom_wts, color="blue")
ax.set_xlim(-10, 10)
plt.show()

H_comp = Hlmbda_3(lmbda_3.quad_pts)
H_analytic = Hlmbda_3_analytic(lmbda_3.quad_pts)
rel_err = np.linalg.norm(H_comp - H_analytic) / np.linalg.norm(H_analytic)

fig, ax = plt.subplots(figsize=(5, 5))
ax.set_title(f"Hilbert Transform $H[\lambda_3]$ (Rel. Err. {rel_err:.2e})")
ax.plot(lmbda_3.quad_pts, H_comp, color="blue")
ax.plot(lmbda_3.quad_pts, H_analytic, color="k", linestyle='dashed')
ax.set_xlim(-20, 20)
#ax.set_yscale("symlog")
ax.legend(["Computed", "Analytic"])
plt.show()
```
