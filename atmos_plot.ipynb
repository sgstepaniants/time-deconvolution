```python
import math
import numpy as np
import scipy
import matplotlib.pyplot as plt

plt.rc('text', usetex=True)
plt.rc('font', family='serif', size=8)
plt.rcParams.update({
    'font.size': 8,
    'text.usetex': True,
    'text.latex.preamble': r"\usepackage{amsfonts}" + "\n" + r"\usepackage{amsmath}" + "\n" + r"\usepackage{amssymb}"
})
```

```python
atmos_data = scipy.io.loadmat("atmos_data.mat")
```

```python
t = atmos_data['atmos_dat'][:, 0]
atmos_data_true = atmos_data['atmos_dat'][:, 1]
atmos_data_pred = atmos_data['candidate'][:, 0]

plt.plot(t, atmos_data_true)
plt.plot(t, atmos_data_pred)
plt.show()

betas = atmos_data["ker_inv"][:, 0]
alphas = atmos_data["ker_inv"][:, 1]
J = np.exp(-t[:, None]*alphas) @ betas

plt.plot(t, J)
plt.show()
```
